```python
import json
from conllu import parse_incr

# Path to your .conllu file
input_file = "TP1/fr_sequoia-ud-test.conllu"
output_file = "sequoia.test.json"

data = []

with open(input_file, "r", encoding="utf-8") as f:
    for sentence in parse_incr(f):
        tokens = [t["form"] for t in sentence]       # words
        pos = [t["upostag"] for t in sentence]       # POS tags
        data.append({"tokens": tokens, "pos": pos})

# Save as JSON
with open(output_file, "w", encoding="utf-8") as out:
    json.dump(data, out, ensure_ascii=False, indent=2)

print(f"✅ Saved {output_file}")
```
